```python
import pandas as pd
import numpy as np
```

```python
cpi_events = pd.read_csv("../data/cpi_events.csv")
cpi_events = cpi_events.rename(columns={
    "date": "date",
    "CPIAUCSL": "event"
})
cpi_events["date"] = pd.to_datetime(cpi_events["date"])
cpi_events["type"] = "CPI"

# Rate
rate_events = pd.read_csv("../data/rate_events.csv")
rate_events = rate_events.rename(columns={
    "date": "date",
    "FEDFUNDS": "event"
})
rate_events["date"] = pd.to_datetime(rate_events["date"])
rate_events["type"] = "Rate"

cpi_events.head()
rate_events.head()
```

```python
dates = pd.date_range(start="2021-01-01", end="2025-12-31", freq="MS")

nfp_dates = []

for d in dates:
    first_friday = d + pd.offsets.Week(weekday=4)
    nfp_dates.append(first_friday)

nfp = pd.DataFrame({
    "date": nfp_dates,
    "event": "NFP Release",
    "type": "NFP"
})

nfp.head()
```

```python
events = pd.concat([cpi_events, rate_events, nfp], ignore_index=True)
events = events.sort_values("date").reset_index(drop=True)

events.head(10)
```

```python
events.to_csv("../data/events_all.csv", index=False)
```
